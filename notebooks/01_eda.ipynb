# Exploratory Data Analysis — Telco Customer Churn


### Objective

Perform exploratory data analysis on the Telco Customer Churn dataset to:
- Understand customer demographics
- Identify churn patterns
- Explore service usage behavior
- Detect correlations and trends that influence churn


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df = pd.read_csv("../data/raw/telco_churn.csv")
df.head()
```

## Missing Values Analysis

```python
plt.figure(figsize=(10,5))
sns.heatmap(df.isna(), cbar=False)
plt.title("Missing Value Heatmap")
plt.show()
```

## Churn Distribution


```python
df['Churn'].value_counts()
```

```python
sns.countplot(data=df, x='Churn')
plt.title("Churn Distribution")
plt.show()
```

## Insights from Churn Distribution

- The dataset is imbalanced: a significantly smaller proportion of customers have churned.
- This imbalance will impact model training — we will need techniques like:
  - class weighting
  - SMOTE
  - balanced sampling

Business Interpretation:
- Churn is a relatively rare event, but each churned customer represents revenue loss.
- The company should focus on targeted retention strategies for the segment that is leaving.


## Categorical Features vs Churn


```python
cols = ['gender', 'Partner', 'Dependents', 'Contract']

plt.figure(figsize=(12,8))
for i, col in enumerate(cols, 1):
    plt.subplot(2, 2, i)
    sns.countplot(data=df, x=col, hue='Churn')
    plt.title(f"{col} vs Churn")
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

## Key Categorical Insights

### Gender
- Churn rates are almost identical for male and female customers.
- Gender is NOT a strong predictor of churn.

### Partner / Dependents
- Customers without partners or dependents churn more.
- Customers with family responsibilities may have more stability and long-term subscriptions.

### Contract Type
- Month-to-month customers have the highest churn rate.
- Two-year contracts have extremely low churn.

Business Conclusion:
- **Long-term contracts reduce churn drastically.**
- Retention strategies should encourage customers to move from month-to-month to annual plans.


## Numeric Feature Distributions


```python
num_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']

df[num_cols].hist(bins=30, figsize=(12,5))
plt.tight_layout()
plt.show()
```

## Numeric Feature Insights

### Tenure
- Tenure shows the strongest churn pattern.
- New customers (low tenure) churn the most.
- Customers who stay longer are far less likely to churn.

### Monthly Charges
- Customers with higher monthly bills show slightly higher churn.
- This suggests price sensitivity among certain customer segments.

### Total Charges
- Lower total charges correlate strongly with higher churn.
- This is expected because low charges usually mean the customer is new.


## Correlation Heatmap


```python
# Select only numeric columns
numeric_df = df.copy()

numeric_df['TotalCharges'] = pd.to_numeric(numeric_df['TotalCharges'], errors='coerce')

numeric_df = numeric_df.select_dtypes(include=[np.number])

plt.figure(figsize=(12,8))
sns.heatmap(numeric_df.corr(), annot=False, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

## Correlation Insights

- Tenure has a strong negative correlation with churn.
  → The longer a customer stays, the less likely they are to leave.

- MonthlyCharges has a slight positive correlation with churn.
  → High monthly bills may contribute to dissatisfaction.

- TotalCharges highly correlates with tenure.
  → This is expected since long-term customers accumulate higher total charges.

Conclusion:
- Tenure is the most powerful single indicator of churn.
- The heatmap confirms the patterns seen in the categorical and numeric analysis.
